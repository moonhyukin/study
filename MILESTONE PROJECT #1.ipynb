```python
# print out a board

def display_board(board):
    print(board[0]+'|'+board[1]+'|'+board[2])
    print("---------")
    print(board[3]+'|'+board[4]+'|'+board[5])
    print("---------")
    print(board[6]+'|'+board[7]+'|'+board[8])
```

```python
# assign player's markers X or O

from IPython.display import clear_output

def player_input():
    
    marker = "wrong"
    
    while marker not in ["X","O"]:
        marker = input("Player 1, choose X or O: ")
        
        if marker not in ["X","O"]:
            clear_output()
            print("invalid option")
            
    player1 = marker
    
    if player1 == "X":
        player2 = "O"
    else:
        player2 = "X"
        
    return (player1,player2)
```

```python
# assign marker into the board

def place_marker(board, marker, position):
    board[position - 1] = marker
```

```python
# takes in a mark, check if win

def win_check(board, mark):
    
    win = False
    
    if board[0] and board[1] and board[2] == mark:
        win = True
    elif board[3] and board[4] and board[5] == mark:
        win = True
    elif board[6] and board[7] and board[8] == mark:
        win = True
    elif board[0] and board[3] and board[6] == mark:
        win = True
    elif board[1] and board[4] and board[7] == mark:
        win = True
    elif board[2] and board[5] and board[8] == mark:
        win = True
    elif board[0] and board[4] and board[8] == mark:
        win = True
    elif board[2] and board[4] and board[6] == mark:
        win = True
    else:
        win = False
        
    return win
```

```python
# who plays first?

import random

def choose_first():
    players = ['X','O']
    
    return random.choices(players)
```

```python
# is space on board available?

def space_check(board, position):
    return board[position-1] == ' '
```

```python
# check if board full

def full_board_check(board):
    for nums in board:
        if nums != "X" and nums != "O":
            return False
        else:
            return True
```

```python
# ask player's next position, check if available, then return position

def player_choice(board):
    
    player_position = int(input("pick your position(1-9): "))
    
    if space_check(board, player_position) == True:
        return player_position
```

```python
# ask if want to replay

def replay():
    
    replay = "wrong"
    
    while replay not in ["Y","N"]:
        replay = input("wanna replay?(Y or N)")
        
        if replay not in ["Y","N"]:
            print("answer Y or N")
            
    if replay == "Y":
        return True
    else:
        return False
```
